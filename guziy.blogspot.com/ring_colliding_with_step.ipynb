# Physics: Ring sliding onto a step

This problem took me awhile as well. Mainly, its second part took me the longest because I could not get the same answer as was in the book. If you have any ideas about the second part of the problem, please leave me your feedback in the comments below the post.

**Problem statement**
 
 A ring of radius $R$ is sliding with a speed $v$ on a floor and hits a corner of a step of height $h$. Determine the height $H$ to which the ring will bounce in the following 2 cases:

 - (a) No friction between the ring and the step.
 - (b) Friction is present between the ring and the step, but no sliding. 
 
*Note*: There is no rotation before the collision.
 
**Solution**

*Case (a): no friction*

Since there is no friction I assumed that the speed after the collision $\vec{v_1}$ would be the same as before, i.e.:

$$
|\vec{v_1}| = v
$$

This is true, provided there is no additional torques acting on the ring that change angular momentum. Well, here I assume that the net force acting on the ring $\vec{F} = \vec{N} + m\vec{g}$ is pointing to the center of the ring, therefore its torque is 0. The force $\vec{F}$ is a sum of the gravity and the reaction of the step at the point during the collision when the ring does not press on the floor.

The two assumptions above are enough to find the angle $\beta$ between $\vec{v_1}$ and $\vec{v}$ (see Fig. 1):

$$
\beta = 2\alpha
$$

From the figure we can find the $\sin\alpha$ and $\cos\alpha$: 

$$
\cos^2\alpha = \frac{(R-h)^2}{R^2} \rightarrow
\sin^2\alpha = \frac{h(2R - h)}{R^2}
$$


Then the jump hegiht of the ring would follow from the energy conservation:

$$
H_1 = \frac{v_1 ^ 2}{2g}\sin^2 \beta = 
\frac{v_1 ^ 2}{2g}\sin^2 2\alpha = 
\frac{v ^ 2}{g}2\sin^2\alpha\cdot\cos^2\alpha = 
2\frac{v ^ 2}{g}\frac{h(2R-h)(R-h)^2}{R^4}
$$

Therefore, the answer in the case without friction is:

$$
H_1 = 2\frac{v ^ 2}{g}\frac{h(2R-h)(R-h)^2}{R^4}
$$

The result I got here is the same as in the problem book.

*Case (b): friction without sliding*

In this case, I have no idea why my answer is different from the one in the book... It might be that I miss something.

Here I assume that the friction force rotates the ring around its center $C$. 

Then, if we denote the point of the ring that touches the step as $O$, it is convenient to consider two components of the ring velocity $\vec{v_{net}}$ after the collision:

$$
\vec{v_{net}} = \vec{v_1} + \vec{v_2}
$$

where $\vec{v_1}\perp OC$ and $\vec{v_2} \parallel OC$.

Then the total jump height can be expressed as:

$$
H_2 = \frac{(v_1\sin\beta_1 + v_2\sin\beta_2)^2}{2g}
$$
