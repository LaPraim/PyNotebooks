```python
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))
b = Basemap(llcrnrlon=-130, urcrnrlon=-100, llcrnrlat=35, urcrnrlat=55, resolution="i")
b.drawcoastlines()
lines = b.drawrivers(color="b")
```

```python
from shapely.geometry import LineString
for p in lines.get_paths():
    line = LineString(p.vertices)
    print(line.wkt)
```
