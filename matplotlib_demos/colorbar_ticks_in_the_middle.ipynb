```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.ticker import FixedLocator
```

```python
x = np.random.randn(10, 10)
```

```python
im = plt.contourf(x, 10)


cb = plt.colorbar(im)
ax = cb.ax

tick_locs = cb.ax.yaxis.get_ticklocs()

loc = ax.yaxis.get_major_locator()


def get_new_locs(old_ticks, ncolors):
    color_height = 1.0 / ncolors
    shift = color_height / 2.0

    new_ticks = old_ticks + shift
    new_ticks = new_ticks[new_ticks <= 1]
    
    return new_ticks
    
loc.locs = get_new_locs(loc.locs, len(cb.boundaries) - 1)

```
