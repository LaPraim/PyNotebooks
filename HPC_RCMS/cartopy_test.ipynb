```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy
from cartopy import crs
from matplotlib.font_manager import FontProperties
```

```python
data = {
    "Guillimin": (45.5017, -73.5673),
    "Graham": (43.4643, -80.5204),
    "Cedar\nOrcinus": (49.2827, -123.1207),
    "Colosse": (46.8139, -71.2080)
}
```

```python
fig = plt.figure(figsize=(6, 6), dpi=300)

rpole = cartopy.crs.RotatedPole(central_rotated_longitude=-100, pole_latitude=55, pole_longitude=60)

ax = plt.axes(projection=rpole)



import cartopy.feature as cfeature
land_50m = cfeature.NaturalEarthFeature('physical', 'land', '50m',
                                        edgecolor='k',
                                        facecolor=cfeature.COLORS['land'], linewidth=0.3)

#coastline_50m = cfeature.NaturalEarthFeature('physical', 'coastline', '50m')


#ax.add_feature(cartopy.feature.LAND)
ax.add_feature(land_50m)

ax.add_feature(cartopy.feature.OCEAN)
#ax.add_feature(cartopy.feature.COASTLINE)
# ax.add_feature(coastline_50m)
ax.add_feature(cartopy.feature.BORDERS, linestyle='-', linewidth=0.5)
ax.add_feature(cartopy.feature.LAKES, alpha=0.5)
#ax.add_feature(cartopy.feature.RIVERS)
provinces_50m = cfeature.NaturalEarthFeature('cultural',
                                             'admin_1_states_provinces_lines',
                                             '50m',
                                             facecolor='none', edgecolor="k", linewidth=0.2
                                            )
ax.add_feature(provinces_50m)

ax.set_extent([-140, -50, 45, 80])



start = data["Guillimin"]

annotate_transform = crs.Geodetic()._as_mpl_transform(ax)
bbox_props = dict(fc="w", alpha=0.7, ec="b", lw=0.1, boxstyle="square,pad=0.2")

for label, finish in data.items():
    #ax.plot([start[1], finish[1]], [start[0], finish[0]], c="b", transform=rpole, lw=2)
    ax.plot([start[1], finish[1]], [start[0], finish[0]], c="r", transform=crs.Geodetic(), lw=1, alpha=0.8)

    ax.scatter(finish[1], finish[0], marker="*", c="orange", s=50,  transform=crs.Geodetic(), zorder=10, linewidth=0.1)
    
    ax.annotate(label, xy=(finish[1] + 1.5, finish[0]), ha="left", va="top", 
                xycoords=annotate_transform, zorder=11, 
                fontproperties=FontProperties(size=5), bbox=bbox_props)


plt.savefig("hpc_connections.png", transparent=True, dpi=300, bbox_inches="tight")
```
