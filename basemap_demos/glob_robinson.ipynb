```python
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.basemap import shiftgrid
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import maskoceans
```

```python
bm = Basemap(projection='robin', resolution = 'l', area_thresh = 1000.0,  lat_0=0, lon_0=0)
```

```python
plt.figure(figsize=(8, 8))

nx = 100
ny = 100
lons = np.linspace(0, 359.9, nx)
lats = np.linspace(-90, 89.9, ny)
lons, lats = np.meshgrid(lons, lats)
data = lons

# plotting
lons[lons >= 180] -= 360

lons, data = bm.shiftdata(lons, datain=data, lon_0=0)
data = maskoceans(lons, lats, data)

xx, yy = bm(lons, lats)

bm.drawcoastlines(linewidth=0.3)
im = bm.pcolormesh(xx, yy, data[:-1,:])

bm.colorbar(im)
```
