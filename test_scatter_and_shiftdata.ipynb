```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python
lons = list(np.arange(-180, -140, 1)) + list(np.arange(140, 185, 1))
lats = np.linspace(-20, 20, len(lons))
```

```python
lats
```

```python
b = Basemap()
```

```python
b.drawcoastlines()
b.scatter(lons[0], lats[0], latlon=True)
```

```python

b.drawcoastlines()

#Scatter
b.scatter(lons, lats, latlon=True)

lat2d, lon2d = np.meshgrid(lats, lons)

data = np.random.randn(*lon2d.shape)
b.contourf(lon2d, lat2d, data, latlon=True, zorder=-1)
```
