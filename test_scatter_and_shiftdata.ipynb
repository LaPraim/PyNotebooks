```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python
lons = list(np.arange(170, 355, 1))
lats = np.linspace(-20, 20, len(lons))
```

```python
b = Basemap(urcrnrlon=400)
```

```python
b.drawcoastlines()
b.scatter(lons[0], lats[0], latlon=True)
```

```python

# with shiftdata

b.drawcoastlines()

#Scatter
lons1 = b.shiftdata(lons)
b.scatter(lons1, lats, latlon=True)

lon2d, lat2d = np.meshgrid(lons, lats)

data = np.random.randn(*lon2d.shape)
b.pcolormesh(lon2d, lat2d, data, latlon=True, zorder=-1)
```

```python
# without shiftdata (but with an adjustment of longitude for scatter)

plt.figure(figsize=(20, 10))

b.drawcoastlines()

#Scatter
b.scatter(lons, lats, latlon=True)

lon2d, lat2d = np.meshgrid(lons, lats)

data = np.random.randn(*lon2d.shape)
b.pcolormesh(lon2d, lat2d, data, latlon=True, zorder=-1)
```
