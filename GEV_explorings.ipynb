#### Journal of Climate. A paper by F. Zwiers and V. Kharin. 1997: "Changes in Extremes of the Climate Simulated by CCC GCM2 under CO2 doubling".

Typing errors which I've detected:

* **p2202**, error in the ranges of $x$ in the expression for $F(x)$ (i.e. CDF), for $k\gt0$ and for $k\lt0$. 

    Should probably be 
    
    $$
\left\\{
\begin{eqnarray}
    k \lt 0 \rightarrow  x \gt \zeta + \alpha/k \\\\
    k \lt 0 \rightarrow x \lt \zeta + \alpha/k
\end{eqnarray}
\right.
    $$

    Anyway something should be changed in the formula. If one considers a function $f(x) = 1 - k(x-\zeta)/\alpha$.
    Then if the condition $x \gt \zeta + \alpha/k$ for $k>0$ is true, then

    $$
    f(x) \lt 1 - k(\alpha/k + \zeta - \zeta)/\alpha = 0 \rightarrow f(x) \lt 0
    $$

    Which would imply a negative number under a rational power in the expression for $F(x)$, which is not defined.

    Another way to confirm the problem is to put numbers into conditions from the paper (I took the numbers from the examples in figure 1 in the article):

    $$
    k \gt 0 \rightarrow x \gt \zeta + \alpha/k = 0 + 1/0.2 = 5 \rightarrow x \gt 5
    $$

    For the negative shape parameter values:
    
    $$
    k \lt 0 \rightarrow x \lt \zeta + \alpha/k = 0 + 1/(-0.2) = -5 \rightarrow x \lt -5
    $$
    
    Which even do not overlap with the ranges in the figure 1 in the paper.   

* **p2202**, in the figure label, should be scale parameter $\alpha=1 $, not 0.

So the modified formula in the article should look as follows (just for my reference):

$$
F(x)=\left\\{
\begin{eqnarray}
\exp\left\\{ -\left[ 1 - k(x-\zeta)/\alpha \right]^{1/k} \right\\}, & k \lt 0, & x \gt \zeta + \alpha/k, \\\\
\exp\left\\{ -\exp\left[ 1 - k(x-\zeta)/\alpha \right] \right\\},   & k = 0,   &   \\\\
\exp\left\\{ -\left[ 1 - k(x-\zeta)/\alpha \right]^{1/k} \right\\}, & k \lt 0, & x \lt \zeta + \alpha/k.
\end{eqnarray}
\right.
$$

#### Playing around with the formulas:


```python
import numpy as np

def gev(x, zeta, k, alpha):
    if not k:
        return np.exp(-np.exp(-(x - zeta)/alpha))
    return np.exp(-(1-k*(x - zeta)/alpha)**(1.0/k))

def gev_pdf(x, zeta, k, alpha):
    if not k:
        t = np.exp(-(x-zeta)/alpha)
    else:
        t = (1-k*(x-zeta)/alpha)**(1.0/k)
    return 1.0/alpha * t**(-k+1)*np.exp(-t)


x = np.arange(-2, 5, 0.1)

###Plot the graphs
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

gs = GridSpec(1,3)
fig = plt.figure(figsize=(20,6))

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[0,2])

k_list = [-0.2, 0, 0.2]
line_width=1.5
for k in k_list:
    y = gev(x, 0, k, 1.0)
    ax1.plot(x, y, "-", label="k={0}".format(k), lw=line_width) 
    ax2.plot(x, gev_pdf(x, 0, k, 1.0), "-", label="k={0}".format(k), lw=line_width)
    
    ax3.plot(-np.log(-np.log(y)), x, label = "k={0}".format(k), lw=line_width)


from matplotlib.ticker import MultipleLocator
    
ax1.set_xlabel("x")
ax1.set_ylabel("CDF")

ax2.set_xlabel("x")
ax2.set_ylabel("PDF")
    
ax3.set_xlabel("Y (Gumbel variate)")
ax3.set_ylabel("x")
ax3.set_xlim([-2,5])
ax3.xaxis.set_minor_locator(MultipleLocator(base=1))

    
ax4 = ax3.twiny()
ax4.set_xlim(ax3.get_xlim())
ax4.set_xlabel("Return Period: $T = 1/(1-F)$\n")
ret_periods = [5, 10, 20, 50, 100]
ys = [-np.log(-np.log(1.0 - 1.0/T)) for T in ret_periods ]
ax4.set_xticks(ys)
ax4.set_xticklabels(ret_periods)
ax4.grid("on")


    
ax1.legend(loc=4)
ax2.legend()
ax3.legend(loc=2)
```
