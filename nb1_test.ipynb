```python
from matplotlib import gridspec
from matplotlib import cm


fig = plt.figure()
nrows = 2
ncols = 2
gs = gridspec.GridSpec(nrows,ncols + 1, wspace = 0.2, hspace=0.4, width_ratios=[1,1,1] )

cmap = cm.get_cmap("cool", 10)

for i in range(nrows):
    for j in range(ncols):
        ax = fig.add_subplot(gs[i,j])
        ax.set_title("({0},{1})".format(i, j))
        x = np.linspace(0, 1, 100)
        y = x
        y, x = np.meshgrid(y, x)
        z = x**2 + y**2 + i * x + j * y
        img = ax.imshow(z,  vmin = 0, vmax = 3, cmap = cmap)
        ax.set_anchor("W")
ax = fig.add_subplot(gs[:, 2])
ax.set_aspect(20)
ax.set_anchor("W")
fig.colorbar(img, cax = ax)
#fig.tight_layout()
```
