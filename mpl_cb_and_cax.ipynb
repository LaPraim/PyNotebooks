```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np
```

```python
gs = GridSpec(1,2, width_ratios=[1, 0.05], wspace=0)
```

```python
nx = 10
ny = 10
data = np.random.randn(nx, ny)
```

```python
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(gs[0, 0])
cs = ax.contourf(data)

cax = fig.add_subplot(gs[0, 1])
plt.colorbar(cs, cax=cax)
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
b = Basemap(lon_0=0)
lats, lons = np.meshgrid(np.linspace(-20,20,ny), np.linspace(-20, 20, nx))
xx, yy = b(lons, lats)
```

```python
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(gs[0, 0])
cs = b.contourf(xx, yy, data)
b.drawcoastlines()
ax.set_aspect("auto")

cax = fig.add_subplot(gs[0, 1])
plt.colorbar(cs, cax=cax)
```
