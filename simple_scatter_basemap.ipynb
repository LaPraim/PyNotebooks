```python
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(15, 10))
m = Basemap(resolution ='c')
m.drawcoastlines()
m.drawcountries()
m.drawmeridians(np.arange(0, 360, 30))
m.drawparallels(np.arange(-90, 90, 30))
m.bluemarble()

lon_list = np.linspace(-180, 180, 60)
lat_list = np.linspace(-90, 90, len(lon_list))
data = np.random.randn(len(lon_list))

x, y = m(lon_list, lat_list)
# You can also set the size according to your data
s = m.scatter(x, y, c=data, s=40)
m.colorbar(s)
```
