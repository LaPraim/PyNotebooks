```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ecoclimap_albedos(sand, orgm=0):
    algdry = 0.10+(0.05 + max(0.0030*sand,0.1))*((1-0.9*orgm**2))
    return algdry, algdry / 2.0
```

```python
def glc2000_albedos(sand):
    algdry = min(0.14+0.0046*sand, 0.45)
    algwet = 0.08+0.0006*sand
    return algdry, algwet
```

```python
ds = 0.1
s = np.arange(0, 100 + ds, ds)

before = np.vectorize(ecoclimap_albedos)(s)
after = np.vectorize(glc2000_albedos)(s)
```

```python
# wet albedo
plt.title("WET")
plt.plot(s, before[1], label="Before")
plt.plot(s, after[1], label="After")
plt.legend(loc="upper left")
```

```python
# dry albedo
plt.title("DRY")
plt.plot(s, before[0], label="Before")
plt.plot(s, after[0], label="After")
plt.ylim([0, 0.5])
plt.legend(loc="upper left")
```
