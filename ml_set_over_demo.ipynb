```python
data = np.random.randn(20, 20)
data.min(), data.max()
```

```python
cs = plt.contourf(data, levels=np.arange(-1, 1, 0.1), extend="both")
cs.set_clim(cs.cvalues[1], cs.cvalues[-2])
plt.colorbar()
print cs.cvalues
```
