```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#defaults
plt.title("Test")
plt.plot(x, y)
```

```python
from matplotlib import rc
from matplotlib.font_manager import FontProperties
font = {'family' : 'monospace',
        'weight' : 'bold',
        'size'   : 20}

rc('font', **font)  # pass in the font dict as kwargs

plt.title("Test", fontproperties=FontProperties(**font)) # to customize your title
plt.plot(x, y)
```
