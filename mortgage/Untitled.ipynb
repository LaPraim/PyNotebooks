```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
def monthly_payment(borrow, nmonth_total, rate_monthly):
    x = (rate_monthly + 1) ** nmonth_total
    return borrow * rate_monthly * (x / (x - 1))

def get_interest_for_month(borrow, kmonth, rate_monthly, nmonth_total):
    x = (rate_monthly + 1) ** (kmonth - 1)
    return x * borrow * rate_monthly + monthly_payment(borrow, nmonth_total, rate_monthly) * (1 - x) 
    
```

```python
plt.figure(figsize=(6, 3), dpi=100)
total_years = 25
total_months = total_years * 12
house_price = 290000
# downpayment = 0.2 * house_price
downpayment = 30000
insurance = 14000
borrow = (house_price - downpayment + insurance)
rate_monthly = 2.99 / 12 / 100
m = np.arange(1, total_months + 1, 1)
plt.plot(m, get_interest_for_month(borrow, m, rate_monthly, total_months))
plt.grid(True)

print(f"Total should be paid to the bank: $ {monthly_payment(borrow, total_months, rate_monthly) * total_months + downpayment}")
print(f"Monthly payment: $ {monthly_payment(borrow, total_months, rate_monthly)} for {total_months} months")


plt.figure(figsize=(6, 3), dpi=100)
m = np.arange(1, 5 * 12 + 1, 1)
plt.plot(m, get_interest_for_month(borrow, m, rate_monthly, total_months))
plt.grid(True)


print(f"Total for the first 5 years: {5 * 12 * monthly_payment(borrow, total_months, rate_monthly) + downpayment}")
```
