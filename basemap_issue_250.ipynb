```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python
ny = 20
nx = 20
yy, xx = np.meshgrid(np.linspace(-50, 50, ny), np.linspace(-50,50, nx))
b = Basemap(projection="cyl")
```

```python
z = xx[:-1,:-1] ** 3 + yy[:-1,:-1] ** 3 
b.pcolormesh(xx, yy, z, latlon=True)
b.drawcoastlines()
```
