```python
folder_path = "/home/huziy/skynet3_rech1/dfs_data_for_cnrcwp_tutorial_3dec_2014"
```

```python
import os
from netCDF4 import Dataset
from mpl_toolkits.basemap import Basemap, maskoceans
import numpy as np
```

```python
os.listdir(folder_path)
```

```python
filename = 'drowned_radsw_DFS5.2_y2012.nc'
filepath = os.path.join(folder_path, filename)
```

```python
filepath
```

```python
ds  = Dataset(filepath)
```

```python
ds.variables.keys()
```

```python
radsw_var = ds.variables["radsw"]
lon_var, lat_var = ds.variables["lon0"], ds.variables["lat0"]
lon2d, lat2d = np.meshgrid(lon_var, lat_var)
```

```python
radsw_var.shape, lon_var.shape
```

```python
b = Basemap(projection="sinu",lon_0 = 180)
```

```python
x, y = b(lon2d, lat2d)
```

```python
plt.figure(figsize=(10, 10))

lons2d_1 = lon2d[:,:]
lons2d_1[lons2d_1 > 180] -= 360
data_masked_ocean = maskoceans(lons2d_1, lat2d, radsw_var[15])

data_masked_land = np.ma.masked_where(~data_masked_ocean.mask, radsw_var[15])

im = b.contourf(x, y, data_masked_land)
b.drawcoastlines()
b.colorbar(im)
plt.title(radsw_var.units)
```

```python
import pandas as pd
from netCDF4 import num2date
```

```python
radsw_data = radsw_var[:]
```

```python
time_num = ds.variables["time"][:]
```

```python
time_num[:10]
```

```python
dates = num2date(time_num, ds.variables["time"].units)
```

```python
dates[:10]
```

```python
panel = pd.Panel(data=radsw_data, items=dates, major_axis=range(radsw_data.shape[1]), minor_axis=range(radsw_data.shape[2]))
```

```python
panel
```

```python
sm = panel.groupby(lambda d: d.month // 3 if d.month < 12 else 0, axis='items').mean()
```

```python
def plot_for_season(index):
    plt.figure(figsize=(10, 10))

    lons2d_1 = lon2d[:,:]
    lons2d_1[lons2d_1 > 180] -= 360
    data_masked_ocean = maskoceans(lons2d_1, lat2d, radsw_var[15])

    data_masked_land = np.ma.masked_where(~data_masked_ocean.mask, sm[index])

    im = b.contourf(x, y, data_masked_land)
    b.drawcoastlines()
    b.colorbar(im)
    plt.title(radsw_var.units)
```

```python
plot_for_season(0)
```

```python
plot_for_season(1)
```

```python
plot_for_season(2)
```
