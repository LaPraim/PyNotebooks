```python
%matplotlib inline

from scipy.stats import gamma
import matplotlib.pyplot as plt
import os
import numpy as np

fig = plt.figure(figsize=(10, 10))


# take the data
data = np.loadtxt(os.path.expanduser('~/Downloads/test.dat'),skiprows=1) 
n, bins, patches = plt.hist(data[:,1],bins=50, normed=1, visible=True,color='white')
col=['red','magenta','green', "k"]

#estimate gamma params
param = gamma.fit(data[:,1],floc=0) 
print(param)

i=0
for val in [500,800,1000, 10000]:
    x = np.linspace( np.min(data[:,1]) , np.max(data[:,1]), val)
    # create a fitted curve
    pdf_fitted = gamma.pdf( x, param[0], scale=param[-1] ) 
    plt.plot(x, pdf_fitted, c=col[i],lw=2, label=str(val)) 
    i += 1

plt.legend()
plt.show()
```
